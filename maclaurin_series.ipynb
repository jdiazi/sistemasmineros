```python
import math
import numpy as np
import pandas as pd
```

#### PASO 1: Definiendo constante $c=\frac{1}{\sqrt{2\pi}}$ y vector nulo con 100 entradas

```python
# c: constante
c = 1/math.sqrt(2*math.pi)
# n: vector nulo con 100 valores
n = np.zeros(100)
```

#### PASO 2: Aplicando la serie de Maclaurin a la distribución normal

```python
def normal(x):
    for i in range(len(n)):
        n[i] = c* ((x**(2*i + 1)) * ((-1)**i))/((2*i + 1) * (2**i) * math.factorial(i))  
    return np.sum(n)
```

#### PASO 3: Asignando valores a los encabezados de filas y columnas

```python
# Col = {0, 1, ..., 9}
col = np.arange(0, 10)
# Col = {0.0, 0.1, ..., 3.4}
fil = np.arange(0, 35)/10
```

#### PASO 4: Creando Data frame con valores unitarios temporales

```python
df2 = pd.DataFrame(1, index=fil, columns=col)
```

#### PASO 5: Aplicando la función normal para cada valor según fila y columna

```python
for i in range(10):
    df2[i] = np.array(list(map(normal, fil + i/100)))
```
